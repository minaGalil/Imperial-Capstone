```python
import numpy as np

# Load files
X = np.load("function1_inputs.npy")
y = np.load("function1_outputs.npy")

# Check shapes
print("Inputs shape:", X.shape)
print("Outputs shape:", y.shape)

"""
•	there are 10 tested input points
•	each point has 2 values
•	each point has one corresponding output
"""
#print(X)
#print(y)

# Find best point
# exploitation : use the best-known result as a guide for the next search step
best_index = np.argmax(y)       # index where the output is highest
best_input = X[best_index]      # the input point at that row
best_output = y[best_index]     # the highest output value

print("\nBest index:", best_index)
print("Best input:", best_input)
print("Best output:", best_output)


# 🔒 Fix randomness
np.random.seed(42)
"""
Without this line, every time I get a different random shift.using this help me to reproduce the results
    •	the result stays consistent
    •	my submission is repeatable
    •	my logic is easier to explain
"""

# Generate controlled shift
# Mixing
#  •	exploitation: stay near best point
#  •	exploration: don’t repeat the exact same point
shift = np.random.uniform(-0.05, 0.05, size=len(best_input))

# Create new point
new_point = best_input + shift
new_point = np.clip(new_point, 0, 1)    # Forces all values to stay between 0 and 1.

print("New point:", new_point)

# Format for portal
query_string = "-".join([f"{value:.6f}" for value in new_point])    # Convert the point into portal format -->  0.xxxxx-0.yyyyyy

print("\n FINAL SUBMISSION:")
print(query_string)
```
